```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Read the dataset to python environment

```python
#. Read the dataset to python environment
data=pd.read_csv('iris.csv')
data.head()
```

## Display the columns in the dataset

```python
data.columns
```

## Calculating the mean

```python
mean1=data['SL'].mean()
mean2=data['SW'].mean()
mean3=data['PL'].mean()
mean4=data['PW'].mean()
print(mean1)
print(mean2)
print(mean3)
print(mean4)
```

# visualisations

```python
plt.figure()
plt.hist(data[['SL']],rwidth=0.9)
plt.title('Distribution of SL',fontSize=16)
plt.xticks()
plt.yticks()
plt.xlabel('SL')
plt.ylabel('Frequency')
```

```python
plt.figure()
plt.hist(data[['SW']],rwidth=0.9,color='red')
plt.title('Distribution of SW',fontSize=16)
plt.xticks()
plt.yticks()
plt.xlabel('SW')
plt.ylabel('Frequency')
```

```python
plt.figure()
plt.hist(data[['PL']],rwidth=0.9,color='green')
plt.title('Distribution of PL',fontSize=16)
plt.xticks()
plt.yticks()
plt.xlabel('PL')
plt.ylabel('Frequency')
```

```python
plt.figure()
plt.hist(data[['PW']],rwidth=0.9,color='grey')
plt.title('Distribution of PW',fontSize=16)
plt.xticks()
plt.yticks()
plt.xlabel('PW')
plt.ylabel('Frequency')
```

```python
plt.figure()
data[['SL','SW','PL','PW']].plot.kde()
plt.title('Probability Density',fontSize=16)
plt.xticks()
plt.yticks()
plt.xlabel('Values')
```

```python
plt.figure()
data[['SL','SW','PL','PW']].plot.hist(rwidth=0.9)
plt.title('Frequency Distribution',fontSize=16)
plt.xticks()
plt.yticks()
plt.xlabel('Values')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,1,1)
plt.scatter(data['SL'],data['PL'],s=15,c='red')
plt.title('SL vs PL',fontSize=16)
plt.xticks()
plt.yticks()
plt.xlabel('SL',fontSize=12)
plt.ylabel('PL',fontSize=12)

plt.figure(figsize=(8,8))
plt.subplot(2,1,2)
plt.scatter(data['SW'],data['PW'],s=15,c='blue')
plt.title('SW vs PW',fontSize=16)
plt.xticks()
plt.yticks()
plt.xlabel('SW',fontSize=12)
plt.ylabel('PW',fontSize=12)

plt.tight_layout()
```
