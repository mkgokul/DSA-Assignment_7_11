# bar graph of male vs female buyers participated in the sales.

```python



sns.catplot(x=cars['Buyer Gender'],
                data=cars, kind="count",
                height=6, aspect=.5)
plt.title('\nCount of Male and Female Buyers\n',fontsize=16)
plt.ylabel('Count',fontsize=12)
plt.xlabel('Buyer Gender',fontsize=12)
plt.grid(linestyle='--',linewidth=.8)
```

# Find the top 5 cars based on their sales price

```python
cars2=cars.sort_values('Sale Price',ascending=False)
cars3=cars2[0:5]
cars3
```

```python
print('top 5 cars based on their sales price are\n',cars3['Make'])
```

# The least 5 cars based on their Resell price.

```python
cars4=cars.sort_values('Sale Price',ascending=True)[0:5]

cars4
```
